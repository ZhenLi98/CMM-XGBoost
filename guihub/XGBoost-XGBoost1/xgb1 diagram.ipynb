```python



import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style
from matplotlib import colors
```

```python
df=pd.read_csv('diagram.csv')
df.head()
```

```python
x=df['TRUE']
y=df['fit']
z=df['prediction']
```

```python
plt.figure(figsize=(50, 25))
plt.rcParams.update({'font.size': 30})
plt.plot(range(len(x)), x, label='true', lineWidth=5)
plt.plot(range(len(y)), y, label='fit', lineWidth=5)
plt.plot(range(len(z)), z, label='prediction', lineWidth=5)
plt.ylabel('U/m')
plt.xlabel('Time')
plt.legend()
plt.axvline(730, -1, 1,color="red")#
plt.show()
```
